```python
class Musica:
    def __init__(self, nombre, anio):
        self.nombre = nombre
        self.anio = anio

    def __str__(self):
        return f"{self.nombre} ({self.anio})"


def mostrar_musicas(musicas):
    if musicas:
        print("musicas ingresadas:")
        for musica in musicas:
            print(musica)
    else:
        print("No se han ingresado Musicas.")


def eliminar_musica(musicas):
    if musicas:
        nombre = input("Ingrese el nombre de la musica a eliminar: ")
        musicas_filtradas = [musica for musica in musicas if musica.nombre != nombre]
        if len(musicas_filtradas) < len(musicas):
            musicas[:] = musicas_filtradas
            print("Musica eliminada exitosamente.")
        else:
            print("No se encontró una Musica con ese nombre.")
    else:
        print("No se han ingresado Musicas.")



def modificar_musica(musicas):
    if musicas:
        nombre = input("Ingrese el nombre de la Musica a modificar: ")
        for musica in musicas:
            if musica.nombre == nombre:
                nuevo_nombre = input("Ingrese el nuevo nombre de la Musica: ")
                nuevo_anio = int(input("Ingrese el nuevo año de la Musica: "))
                musica.nombre = nuevo_nombre
                musica.anio = nuevo_anio
                print("Musica modificada exitosamente.")
                return
        print("No se encontró una Musica con ese nombre.")
    else:
        print("No se han ingresado Musicas.")

def menu():
    musicas = []

    while True:
        print("------ MENÚ ------")
        print("1. Ingresar Musicas")
        print("2. Mostrar Musicas")
        print("3. Eliminar Musica")
        print("4. Modificar Musica")
        print("0. Salir")

        opcion = input("Ingrese una opción: ")

        if opcion == "1":
            nombre = input("Ingrese el nombre de la Musica: ")
            anio = int(input("Ingrese el año de la Musica: "))
            musica = Musica(nombre, anio)
            musicas.append(musica)
            print("Musica ingresada exitosamente.")
        elif opcion == "2":
            mostrar_musicas(musicas)
        elif opcion == "3":
            eliminar_musica(musicas)
        elif opcion == "4":
            modificar_musica(musicas)
        elif opcion == "0":
            print("¡Hasta luego!")
            break
        else:
            print("Opción inválida. Por favor, ingrese una opción válida.")


menu()
```
